```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month'] = df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year'] = df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df.head(2)
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Dept_hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dept_min']= df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_hours'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
df.drop('Route', axis=1,inplace=True)
```

```python
# converting duration data into minutes
df = pd.DataFrame(df)
# Extract hours and minutes
df["Hours"] = df["Duration"].str.extract(r"(\d+)h").astype(float)
df["Minutes"] = df["Duration"].str.extract(r"(\d+)m").astype(float)

# Convert to total minutes
df["Total_Duration_Minutes"] = df["Hours"] * 60 + df["Minutes"]
```

```python
df.drop('Duration', axis=1, inplace=True)
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3,'4 stops':4, np.nan:1})
```
